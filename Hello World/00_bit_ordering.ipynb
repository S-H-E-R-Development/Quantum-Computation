```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
```

```python
# Quantum circuit with 2 qubit

qc = QuantumCircuit(2) 
```

```python
# Adding a Hadamard Gate

qc.h(0)

qc.draw('mpl')
```

```python
# let's measure the state of each qubit
# Measure the qubit into the classical bit

# qc.measure(0, 0)
```

```python
# CNOT
qc.cx(0, 1)


qc.draw('mpl')
```
